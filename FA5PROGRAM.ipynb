```python
def calculate_total_price(price_per_item, quantity):
    """
    Calculates the total price of items.

    Args:
        price_per_item (float): The price of a single item.
        quantity (int): The number of items purchased.

    Returns:
        float: The total price.
    """
    return price_per_item * quantity


try:
    item_price = float(input("Enter the price of one item: "))
    num_items = int(input("Enter the number of items purchased: "))


    total_cost = calculate_total_price(item_price, num_items)


    print(f"\nPrice per item: ${item_price:.2f}")
    print(f"Quantity purchased: {num_items}")
    print(f"Total price: ${total_cost:.2f}")
except ValueError:
    print("Invalid input. Please enter valid numbers for price and quantity.")
```
